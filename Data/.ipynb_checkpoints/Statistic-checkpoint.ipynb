```python
def make_statistic_magic(list_of_data):
    izotherm_x_y_copy = copy.deepcopy(list_of_data)
    lenght_of_each_izotherm = []
    for x_y in izotherm_x_y_copy:
        lenght_of_each_izotherm.append(len(x_y))
    
    data1 = pd.Series(lenght_of_each_izotherm);
     
    value_counts = data1.value_counts()
    
    plt.bar(value_counts.index, value_counts.values)
    
    plt.title('Wykres częstości ilości punktów pomiarowych w izotermie')
    plt.xlabel('Ilość punktów pomiarowych w izotermie')
    plt.ylabel('Częstość')
    
    
    mean_freq = data1.mean()  
    median_freq = data1.median()  
    std_dev_freq = data1.std()  
    first_quartile = data1.quantile(0.25)
    third_quartile = data1.quantile(0.75)
    min_length = data1.min()
    max_length = data1.max()
    
    print(f"Statystyki liczby punktów w izotermie na podstawie {len(izotherm_x_y_copy)} próbek: ")
    print("Średnia :", mean_freq)
    print("Mediana :", median_freq)
    print("Odchylenie standardowe :", std_dev_freq)
    print("Pierwszy kwartyl :", first_quartile)
    print("Trzeci kwartyl :", third_quartile)
    print("Min:", min_length)
    print("Maks:", max_length)
    
    
    plt.show()
make_statistic_magic(oryginal_isotherms)
```
